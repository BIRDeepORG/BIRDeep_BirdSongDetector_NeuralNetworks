```python
import os
import pandas as pd
import shutil
```

```python
# Directory paths
ROOT_PATH = "../../../desarrollo/"
SOURCE_PATH = os.path.join(ROOT_PATH, "Audio_Data/")
DESTINATION_PATH = os.path.join(ROOT_PATH, "Data/Dataset/Audios/")

CSV_FILE = os.path.join(ROOT_PATH, "Data/Annotations/audio_annotations.csv")
```

```python
# Load the CSV file
df = pd.read_csv(CSV_FILE)
```

```python
df
```

```python
# Count unique paths in df
unique_paths = df["path"].unique()
print(f"Unique paths: {len(unique_paths)}")
```

```python
moved_files = set()

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Get the source and destination file paths
    source_file_path = os.path.join(SOURCE_PATH, row['path'])
    destination_file_path = os.path.join(DESTINATION_PATH, row['path'].replace('Audios/', ''))

    # Check if the WAV file has already been moved
    if source_file_path in moved_files:
        continue

    # Create the destination directory if it doesn't exist
    os.makedirs(os.path.dirname(destination_file_path), exist_ok=True)

    # Move the file
    # shutil.move(source_file_path, destination_file_path)

    # Update the 'path' value in the DataFrame
    df.at[index, 'path'] = destination_file_path.split("Audios/")[1]

    # Add the file to the moved files set
    moved_files.add(row['path'])

# Save the changes to the DataFrame
df.to_csv(CSV_FILE, index=False)
```

```python
# Show length of moved files
print(f"Moved files: {len(moved_files)}")
```
