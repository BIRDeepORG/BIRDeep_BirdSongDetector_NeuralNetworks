```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Ruta del archivo CSV
csv_path = "../../Data/CSVs/old_network.csv"
```

```python
# Leer el archivo CSV en un DataFrame
df = pd.read_csv(csv_path)
```

```python
# Ruta base para las im√°genes
base_path = "../../Dataset/"
```

```python
# Iterar a trav√©s de las filas del DataFrame y mostrar las im√°genes con detalles
for index, row in df.iterrows():
    image_path = os.path.join(base_path, row["Path"].split("/")[-1])
    class_name = row["Class"]
    prob = row["Prob"]
    
    # Leer y mostrar la imagen
    img = plt.imread(image_path)
    plt.imshow(img)
    
    # Configurar t√≠tulo y subtitulo
    title = os.path.basename(image_path)
    subtitle = f"{class_name} ({prob:.4f})"
    
    plt.title(title + "\n" + subtitle)
    
    plt.show()
```
