```python
PATH = "../../../../desarrollo/"

train_txt = PATH + "Data/TXTs/train.txt"
```

```python
# Take first line which is path of an image, without \n
with open(train_txt, "r") as file:
    image_path = "../../" + file.readline().strip()

# Replace images by labels and .jpg by .txt for labels file
labels_path = image_path.replace("images", "labels")
labels_path = labels_path.replace(".jpg", ".txt")

# Open labels file, each line is a label
with open(labels_path, "r") as file:
    labels = file.readlines()

print(labels)

# Format is: class x_center y_center width height normalized
    

# Load image and plot with Image
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(image_path)
plt.imshow(img)
plt.show()
```
